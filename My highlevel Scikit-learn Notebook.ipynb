# Sk-learn workflow

* end to end sk-learn workflow
* Getting your data ready (to be used with ML models) 
* Choosing a ML model
* Fitting a model to data (learing patterns)
* Making prediction with model(using patterns)
* Evaluating model predictions
* Improving model predictions
* Saving and loading models 

<img src ='sklearn-workflow-title.png'/>

```python
import numpy as np
```

```python
# 1.Getitng your data ready 
import pandas as pd
heart_disease  = pd.read_csv("data_sets/heart-disease.csv")
heart_disease.head()
```

```python
# Create X - features matrix 
X = heart_disease.drop('target',axis=1)

# Create Y - labels 
y = heart_disease['target']
```

```python
# 2. Choosing a ML model and hyperparameters 
from sklearn.ensemble import RandomForestClassifier
clf  = RandomForestClassifier()

#we will keep the default hyperparameters 
clf.get_params()
```

```python
# 3. Fitting a model to  the training data (learing patterns)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test  = train_test_split(X,y,test_size = 0.2) 
```

```python
clf.fit(X_train,y_train);
```

```python
# Making prediction with model(using patterns)
y_preds =  clf.predict(X_test)
y_preds
```

```python
#actual test data
y_test
```

```python
# 4. Evaluating model predictions on training and test data
clf.score(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(classification_report(y_test,y_preds))
```

```python
confusion_matrix(y_test,y_preds)
```

```python
accuracy_score(y_test,y_preds)
```

```python
# 5.Improving model predictions
#Try different amount of n_estimators
np.random.seed(42)
for i in range(10,100,10):
    print(f"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators = i).fit(X_train,y_train)
    print(f"model accuracy on test set: {clf.score(X_test,y_test)*100:.2f}%")
    print("")
```

```python
# 6. Saving and loading models

import pickle 

pickle.dump(clf, open('random_forest_classfier_model_1.pkl','wb'))
```

```python
# our model is saved in .pkl formate by file name "random_forest_classfier_model_1.pkl"
```

```python
# lets load that model back 
loaded_model  = pickle.load(open("random_forest_classfier_model_1.pkl","rb"))
loaded_model.score(X_test,y_test)
```
